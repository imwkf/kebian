迭代方程：$$x_{n+1}=rx_n(1-x_n) $$

定义了一个简单的**确定性**的动力学过程。即只要给定初始值$x_{0}$，就能够确切的知道迭代$n$次后的值$x_{n}$。

### 探索动力学

给某个初值$x_{0}$启动迭代过程，长时间后，$x$会是什么值？请在如下参数下，画出$x$随迭代步数$n$的图像。简单总结一下规律。

$$r=0.5$$
$$r=2$$
$$r=3.3$$
$$r=3.5$$
$$r=3.9$$

```python
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as ipd
#y计算长度 x初值 r的值
def main(x,y,r):
    a=np.zeros((y))
    i=0
    a[0]=x
    while i<y-1:
        a[i+1]=r*a[i]*(1-a[i])
        i=i+1
    b=range(0,y,1)
    plt.plot(b,a,'g')
    plt.xlim(0,y)
    plt.ylim(0,1)
    print('\033[1;31m\tx是初值，y为迭代次数(建议100次)，r为相关参数\033[0m')
ipd.interactive(main,y=[100,1000],x=(0,1,0.05),r=(0,4,0.01))
```

```python
y=100
x=0.3
r=[0.5,2,3.3,3.5,3.9]
plt.figure(figsize=(20,10))
for m in range(5):
    a=np.zeros((y))
    i=0
    a[0]=x
    while i<y-1:
        a[i+1]=r[m]*a[i]*(1-a[i])
        i=i+1
    b=range(0,y,1)
    plt.subplot(3, 2, m+1) #图一包含1行2列子图，当前画在第一行第一列图上
    plt.plot(b,a)
print('                                          \033[1;31m\t以下分别为r等于0.5，2，3.3，3.5，3.9的图像\033[0m')
```

<font color=red>**可以总结出大致规律，随着r的增大，x的取值从少到多，再到趋于混乱**

### 分岔图 （Bifurcation Diagram)

#### 1.研究全参数空间$r∈[0,4]$中，$x$长时间后的行为。

#### 2.在分叉图中，x轴是参数$r∈[0,4]$；$y$轴是经过长时间迭代后$x$的值（去掉暂态之后）。

```python
import matplotlib.pyplot as plt
import numpy as np

r = np.arange(0,4, 0.0001)
x = 0.4  # 初值
plt.figure(figsize=(20,10))
for i in range(200): #全部计算次数
    x = r*x*(1 - x)
    if i >= 100: #100次之后开始计
        #plt.plot(r,x,',g',linewidth=0.1)
        plt.scatter(r,x,c='r',s=0.00001)
        #plt.plot(r,x,',k',alpha=0.05)
plt.show()
```

### 不动点的计算

先定义出所需的函数

```python
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as ipd
x,r=sy.symbols('x r')
f=r*x*(1-x)
ff=r*f*(1-f)
fff=r*ff*(1-ff)
```

求解不动点

```python
X1=sy.solve(f-x,x) #自动等于0
print(X1)
X2=sy.solve(ff-x,x)
print(X2)
```
